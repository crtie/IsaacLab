```python
# z
import matplotlib.pyplot as plt

# 文件路径
file_path = "/home/yudi/Singapore/isaacsim/IsaacLab/my/force_xyz.txt"

# 读取所有行并转成 float
with open(file_path, "r") as f:
    data = [float(line.strip()) for line in f if line.strip()]

# 画折线图
plt.figure(figsize=(12, 6))
plt.plot(data, marker='o', linestyle='-', linewidth=1)
plt.title("Force Z Data")
plt.xlabel("Index")
plt.ylabel("Force Z Value")
plt.grid(True)
plt.show()
```

```python
# x y z xy
import matplotlib.pyplot as plt
import numpy as np

# 文件路径
file_path = "/home/yudi/Singapore/isaacsim/IsaacLab/my/force_xyz_circle.txt"

# 读取所有行，按列解析，适应 f.write(f"{force_x} \t{force_y} \t{force_z} \t{z_pos} \t{x_pos} \t{y_pos}\n") 格式
force_x, force_y, force_z, z_pos, x_pos, y_pos = [], [], [], [], [], []
with open(file_path, "r") as f:
    for line in f:
        if line.strip():
            parts = line.strip().split()  # 自动按空格或制表符分割
            if len(parts) >= 6:
                fx, fy, fz, zp, xp, yp = map(float, parts[:6])
                force_x.append(fx)
                force_y.append(fy)
                force_z.append(fz)
                z_pos.append(zp)
                x_pos.append(xp)
                y_pos.append(yp)
# 转成 numpy
force_x = np.array(force_x)
force_y = np.array(force_y)
force_z = np.array(force_z)
z_pos = np.array(z_pos)
x_pos = np.array(x_pos)
y_pos = np.array(y_pos)
# 计算XY合力 (平面上的合力)
force_xy = np.sqrt(force_x**2 + force_y**2)

# 画图
plt.figure(figsize=(12, 8))
plt.plot(force_x, label='Force X')
plt.plot(force_y, label='Force Y')
plt.plot(force_z, label='Force Z')
plt.plot(force_xy, label='XY resultant force')
plt.plot(z_pos, label='Z position')
plt.plot(x_pos, label='X position')
plt.plot(y_pos, label='Y position')
plt.title("Force Components and Positions Over Time")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(z_pos, label='Z position')
plt.title("Positions Over Time")
plt.xlabel("Index")
plt.ylabel("Position (m)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(x_pos, label='X position')
plt.title("Positions Over Time")
plt.xlabel("Index")
plt.ylabel("Position (m)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(y_pos, label='Y position')
plt.title("Positions Over Time")
plt.xlabel("Index")
plt.ylabel("Position (m)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(force_z, label='Force Z')
plt.title("Force Z Over Time")
plt.xlabel("Index")
plt.ylabel("Force Z (N)")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
plt.plot(force_xy, label='XY resultant force')
plt.title("XY Resultant Force Over Time")
plt.xlabel("Index")
plt.ylabel("Force (N)")
plt.legend()
plt.grid(True)
plt.show()
```
